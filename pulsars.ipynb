```python
def normalize(X):
    d1 = -1.0
    d2 = 1.0
    mini = min(X)
    maxi = max(X)
    return ((X - mini) * (d2 - d1)) / (maxi - mini) + d1
for name in data.columns:
    data[name] = normalize(data[name])
```

```python
y = data['target_class'].values
s = data.columns
X = data[s[:-1]].values
```

```python
from sklearn.model_selection import KFold
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import LogisticRegressionCV
k = KFold(n_splits = 8)
k.get_n_splits(X)
print(k)  
for train, test in k.split(X):
    X_train, X_test, y_train, y_test = X[train], X[test], y[train], y[test]
```

# LOGISTIC REGRESSION

```python
from sklearn.linear_model import LogisticRegression
from sklearn import datasets, linear_model, metrics
from sklearn.metrics import make_scorer, f1_score
```

```python
Cs = [i * 10 for i in range(1, 100)]
penalty = ['l1', 'l2', 'elasticnet', 'none']
params = {
    'C': Cs, 
    'penalty' : penalty
}
```

```python
maxLR = 0
maxC = 0
maxp = ''
_penalty = ['l2', 'none']
for c in Cs:
    for penalty_ in _penalty:
        modelLR = LogisticRegression(C = c, penalty = penalty_)
        modelLR.fit(X_train, y_train)
        y_pred = modelLR.predict(X_test)
        curLR = metrics.f1_score(y_test, y_pred)
        if (maxLR < curLR):
            maxLR = curLR
            maxC = c
            maxp = penalty_
print(maxC, maxp, maxLR)
```

```python
modelLR_R2 = GridSearchCV(LogisticRegression(), param_grid = params, scoring = 'r2')
modelLR_R2.fit(X_train, y_train)
```

```python
modelLR_MSE = GridSearchCV(LogisticRegression(), param_grid = params, scoring = 'neg_mean_squared_error')
modelLR_MSE.fit(X_train, y_train)
```

```python
modelLR_MAE = GridSearchCV(LogisticRegression(), param_grid = params, scoring = 'neg_mean_absolute_error')
modelLR_MAE.fit(X_train, y_train)
```

```python
modelLR_F1 = GridSearchCV(LogisticRegression(), param_grid = params, scoring = 'f1')
modelLR_F1.fit(X_train, y_train)
```

```python
print('R2:')
print(modelLR_R2.best_params_)
print(modelLR_R2.best_score_)
print('MSE:')
print(modelLR_MSE.best_params_)
print(modelLR_MSE.best_score_)
print('MAE:')
print(modelLR_MAE.best_params_)
print(modelLR_MAE.best_score_)
print('F1:')
print(modelLR_F1.best_params_)
print(modelLR_F1.best_score_)
```

# RANDOM FOREST

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import RandomizedSearchCV
```

```python
modelRF = RandomForestClassifier()
n_estimators = [x for x in range(1, 3000, 30)]
max_features = ['log2', 'sqrt']
max_depth = [x for x in range(1, 100, 5)]
min_samples_split = [x for x in range(1, 100, 5)]
min_samples_leaf = [x for x in range(1, 100, 5)]
max_leaf_nodes = [x for x in range(1, 100, 5)]
bootstrap = [True, False]
params = {
            'n_estimators': n_estimators,
            'max_features': max_features,
            'max_depth': max_depth,
            'min_samples_split': min_samples_split,
            'min_samples_leaf': min_samples_leaf,
            'max_leaf_nodes' : max_leaf_nodes,
            'bootstrap': bootstrap
}
modelRS = RandomizedSearchCV(modelRF, 
                        params, 
                        n_iter = 100, 
                        cv = 3, 
                        n_jobs = -1, 
                        random_state = 0,
                        scoring = 'f1')
modelRS.fit(X_train, y_train)
modelRS.best_params_
```
